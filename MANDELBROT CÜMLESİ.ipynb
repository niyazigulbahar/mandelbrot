```python
import numpy as np
import matplotlib.pyplot as plt

# Mandelbrot kümesini çizen fonksiyon
def mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter):
    r1 = np.linspace(xmin, xmax, width)
    r2 = np.linspace(ymin, ymax, height)
    n3 = np.empty((width, height))
    for i in range(width):
        for j in range(height):
            z = 0
            c = r1[i] + 1j * r2[j]
            for k in range(max_iter):
                z = z*z + c
                if abs(z) > 2:
                    break
            n3[i, j] = k
    return (r1, r2, n3)

# Julia kümesini çizen fonksiyon
def julia(xmin, xmax, ymin, ymax, width, height, c, max_iter):
    r1 = np.linspace(xmin, xmax, width)
    r2 = np.linspace(ymin, ymax, height)
    n3 = np.empty((width, height))
    for i in range(width):
        for j in range(height):
            z = r1[i] + 1j * r2[j]
            for k in range(max_iter):
                z = z*z + c
                if abs(z) > 2:
                    break
            n3[i, j] = k
    return (r1, r2, n3)

# Parametreler
width, height = 300, 300
max_iter = 256

# Mandelbrot kümesi çizimi
xmin, xmax, ymin, ymax = -2, 2, -2, 2
r1, r2, mandelbrot_set = mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter)

# Bazı c değerleri seçelim (Mandelbrot kümesinden rastgele)
c_values = [-0.8 + 0.156j, 0.285 + 0j, 0.285 + 0.01j, -0.70176 - 0.3842j]

# Grafik çizimi
fig, axs = plt.subplots(1, len(c_values) + 1, figsize=(20, 5))

# Mandelbrot kümesi
axs[0].imshow(mandelbrot_set.T, extent=[xmin, xmax, ymin, ymax], cmap='hot')
axs[0].set_title('Mandelbrot Kümesi')

# Julia kümeleri
for i, c in enumerate(c_values):
    _, _, julia_set = julia(xmin, xmax, ymin, ymax, width, height, c, max_iter)
    axs[i+1].imshow(julia_set.T, extent=[xmin, xmax, ymin, ymax], cmap='hot')
    axs[i+1].set_title(f'Julia Kümesi\nc = {c}')

plt.tight_layout()
plt.show()
```
